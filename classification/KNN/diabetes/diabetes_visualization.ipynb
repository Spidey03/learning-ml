## Visualization

### Blood Pressure vs Age

```python
g = sns.scatterplot(
    x='Age',
    y='BloodPressure',
    data=df,
    hue="BloodPressure",
    style='Outcome'
)
plt.xlabel('Age')
plt.ylabel('Blood Pressure')
```

## BMI vs Age

```python
sns.catplot(
    x="Age",
    y="BMI",
    data=df,
)

plt.xlabel('Age')
plt.ylabel('BMI')
```

## Blood Pressure vs likely to get Diabetes

```python
sns.catplot(
    y="BloodPressure",
    x="Outcome",
    data=df,
    kind="violin"
)
plt.xlabel("Outcome")
plt.ylabel("BloodPressure")
```
